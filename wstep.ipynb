```python
import requests
import matplotlib.pyplot as plt
import pandas as pd
```

```python
USD=requests.get("https://api.nbp.pl/api/exchangerates/rates/a/usd/2020-01-01/2020-06-16/")

usd_frame = pd.read_json(USD.content)
usd_rates = usd_frame.rates
usd_list=[index['mid'] for index in usd_rates]
plt.plot(usd_list)
plt.xticks(rotation=45)
plt.title("Kurs Dolara")
plt.xlabel("2020-01-01 do 2020-06-16")
plt.ylabel("Cena [Pln]")
plt.legend(["USD"])
plt.grid()

plt.show()
```
